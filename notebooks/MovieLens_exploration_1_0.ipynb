```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
columns_name=['user_id','item_id','rating','timestamp']
df = pd.read_csv("u.data",sep="\t",names=columns_name)
print(len(df))
display(df.head(5))
```

```python
# Visualize the distribution of ratings
plt.figure(figsize=(10, 6))
sns.countplot(x='rating', data=df)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize the average rating per movie
avg_rating_per_movie = df.groupby('item_id')['rating'].mean()
plt.figure(figsize=(15, 6))
avg_rating_per_movie.hist(bins=50, edgecolor='black')
plt.title('Distribution of Average Ratings per Movie')
plt.xlabel('Average Rating')
plt.ylabel('Count')
plt.show()
```

Society did really rate 80+ movies to 1star lfmao

```python
num_ratings_per_user = df.groupby('user_id')['rating'].count()
plt.figure(figsize=(15, 6))
num_ratings_per_user.hist(bins=50, edgecolor='black')
plt.title('Distribution of Number of Ratings per User')
plt.xlabel('Number of Ratings')
plt.ylabel('Count')
plt.show()
```
